```python
alignment_file = pysam.AlignmentFile('../data/MouseDataset/BAMs/BoneMarrow_62016.bam', 'rb')
```

```python
labels = pd.read_csv('./mouse_atlas_cicero/labels.txt', sep='\t', header=None, index_col=0)
```

```python
from collections import defaultdict
```

```python
from tqdm import tqdm
```

```python
def get_read_name(read):
    return read.qname.split(':')[0]
```

```python
OUTPUT_FOLDER = '../data/MouseDataset/SplittedBams'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
reads_by_cells = defaultdict(list)

bam_name = os.path.basename('../data/MouseDataset/BAMs/BoneMarrow_62016.bam')
bam_prefix = os.path.splitext(bam_name)[0]

bam_files = dict()
# pairedreads = pysam.AlignmentFile("allpaired.bam", "wb", template=samfile)


for read in tqdm(alignment_file.fetch('chr1', 5000000, 10000000)):
    print(read.is_paired)
    try:
        cell_type = labels.loc[get_read_name(read)][1]
        cell_type_replaced = cell_type.replace(' ', '_')
        bam_prefix_filename = f'{cell_type_replaced}_{bam_prefix}.bam'
        if bam_prefix_filename not in bam_files:
            typed_reads = pysam.AlignmentFile(bam_prefix_filename, "wb", template=alignment_file)
            bam_files[bam_prefix_filename] = typed_reads
        
        bam_file = bam_files[bam_prefix_filename]
        
        bam_file.write(read)
            
#         print(get_read_name(read))
    except KeyError:
        continue
#     print(read, get_read_name(read))
```

```python
for key, fp in bam_files.items():
    fp.close()
```
