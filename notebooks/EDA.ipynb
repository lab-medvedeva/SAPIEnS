```python
peak_names = pd.read_csv('count_matrix_data/GSE96769_PeakFile.csv')
peak_names = peak_names.set_index('PeakFile_Peak_ID')
```

```python
def get_peaks(cell_name):
    new_df = peak_names.join(df[df[cell_name] > 0], how='inner')[['peak_id', cell_name]].reset_index()
    new_df['cell_name'] = cell_name
    
    new_df = new_df[['peak_id', 'cell_name', cell_name]]
    
    new_df.columns = ['peak_id', 'cell_name', 'counts']
    return new_df
```

```python
peak_names = peak_names.reset_index().set_index('peak_id')
```

```python
peak_names
```

```python
peak_names.set_index('peak_id')
```

```python
from tqdm import tqdm
```

```python
peaks = []

for cell_id in tqdm(range(1, 2954, 1)):
    try:
        peak = get_peaks(f'cell_{cell_id}')
        peaks.append(peak)
    except KeyError:
        continue
```

```python
peaks_df = pd.concat(peaks)
```

```python
peaks_df.to_csv('count_matrix_data/peaks.tsv', sep='\t', header=False, index=None)
```
